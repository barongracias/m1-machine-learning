```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
```

```python
# training points
np.random.seed(280400)
n_points = 50
x_train = np.linspace(0, 1, n_points)
x_train_re = x_train.reshape(-1,1)

# plotting points
x_plot = np.linspace(0, 1, 200)
x_plot_re = x_plot.reshape(-1,1)

# training model
s = np.random.randn(n_points)
y=2*x_train**3 + 10*x_train**2 + s

# plot training data
train_plot = plt.plot(x_train, y, '.', ms=5, label='training')

# poly 3 reg
poly3 = PolynomialFeatures(degree=3)
x3 = poly3.fit_transform(x_train_re)
clf3 = linear_model.LinearRegression()
clf3.fit(x3, y)
x3_plot = poly3.fit_transform(x_plot_re)
poly3_plot = plt.plot(x_plot, clf3.predict(x3_plot), label='poly 3', linewidth=3)

# poly 20 reg
poly20 = PolynomialFeatures(degree=20)
x20 = poly20.fit_transform(x_train_re)
clf20 = linear_model.LinearRegression()
clf20.fit(x20, y)
x20_plot = poly20.fit_transform(x_plot_re)
poly20_plot = plt.plot(x_plot, clf20.predict(x20_plot), label='poly 20', linewidth=3)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Models fit to training data')
plt.legend()
plt.show()
```

```python
# testing points
np.random.seed(280400)
n_test_points = 10
x_test = np.random.random(n_test_points)
x_test_re = x_train.reshape(-1,1)

# plotting points
x_plot = np.linspace(0, 1, 200)
x_plot_re = x_plot.reshape(-1,1)

# testing model
s_test = np.random.randn(n_test_points)
y_test = 2*x_test**3 + 10*x_test**2 + s_test

# plotting points
x_test_plot = np.linspace(0, 1, 200)
x_test_plot_reshape = x_test_plot.reshape(-1,1)
x3 = poly3.fit_transform(x_test_plot_reshape)
x20 = poly20.fit_transform(x_test_plot_reshape)

fig = plt.figure()
p1 = plt.plot(x_test, y_test.transpose(), '.', label='test data', linewidth=5)
p3 = plt.plot(x_test_plot, clf3.predict(x3), label='poly 3', linewidth=2)
p20 = plt.plot(x_test_plot, clf20.predict(x20), label='poly 20', linewidth=2)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Trained models fit to testing data')
plt.legend()
plt.show()
```
